### **Extracting 2018 Movies Data**

```python
import numpy as np
import pandas as pd
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
```

```python
df1=pd.read_html(link, header=0)[2]
df1.head()
```

```python
df2=pd.read_html(link, header=0)[3]
df3=pd.read_html(link, header=0)[4]
df4=pd.read_html(link, header=0)[5]
```

```python
df=df1.append(df2.append(df3.append(df4, ignore_index=True), ignore_index=True), ignore_index=True)
df.head()
```

Now, we observe that above data does not include the genres information. We need to fetch that information from IMDB. We will be using the api key of TmDB to fetch the information of the genres using API.

```python
#We can get API Key here: https://www.themoviedb.org/settings/api
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '3caef0f5a7eb5ec5ed731230f8eccdde'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x): #pass in the title of the movies
    genres = []
    result = tmdb_movie.search(x) #the title will be searched in the tmdb_movie
    movie_id = result[0].id #we will match the "id" with the "title"
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key)) #we will get the result from the IMDb data
    data_json = response.json() #we will then convert it to a json file
    if data_json['genres']: #in the json file we will only need to extract the "genre"
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name']) #we will then add the "genre" to the empty genre list we created above
        return genre_str.join(genres)
    else:
        np.NaN # we will return the results but if we don't find anything we will consider it as a missing value
```

In the above function, we pass the movie title as parameter to the function. We then search for the movie id using the title. Then using the movie id, we fetch the genres name using the api key. We then convert the result into json format. Lastly, we append the genres name with a blank space into an empty list.

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
df.head()
```

```python
#selecting the features of our interest
df_2018 = df[['Title','Cast and crew','genres']]
df_2018
```

Here, we observe that Cast and Crew feature contains the names of the directors and actors. We need to fetch these details into sepearate features for our further analysis. We will be creating functions to fetch these details.

```python
#Getting Director Names
def get_director(x):
    if " (director)" in x: #we are getting "director"
        return x.split(" (director)")[0]
    elif " (directors)" in x: #we are getting "directors" with "s"
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0] #we are getting "directors/screenplay"
```

Here, we are splitting the text by the string (director)/(directors)/(director/screenplay) and fetching the first item from the list.



```python
#let's apply the above function on the "Cast and crew" column
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
df_2018.head()
```

```python
df_2018['Cast and crew'][0]
```

```python
#fethcing the name of Actor-1
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2018['actor_1_name']=df_2018['Cast and crew'].map(lambda x: get_actor1(x))
df_2018.head()
```

```python
#fetching the name of Actor-2
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2018['actor_2_name']=df_2018['Cast and crew'].map(lambda x: get_actor2(x))
df_2018.head()
```

```python
#fetching the name of Actor-2
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2018['actor_3_name']=df_2018['Cast and crew'].map(lambda x: get_actor3(x))
df_2018.head()
```

Renaming the columns as per the namings in the old dataset

```python
df_2018 = df_2018.rename(columns={'Title':'movie_title'})
```

```python
#selecting the relavant columns into a new dataframe
new_df18 = df_2018.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
new_df18.head()
```

```python
new_df18.isna().sum()
```

We need to replace the null values with the unknown string for avoiding errors in further analysis.

```python
new_df18['actor_2_name'] = new_df18['actor_2_name'].replace(np.nan, 'unknown')
new_df18['actor_3_name'] = new_df18['actor_3_name'].replace(np.nan, 'unknown')
```

Now, let's lowercase the character of the movie title feature.

```python
new_df18['movie_title'] = new_df18['movie_title'].str.lower()
```

```python
new_df18['comb'] = new_df18['actor_1_name'] + ' ' + new_df18['actor_2_name'] + ' '+ new_df18['actor_3_name'] + ' '+ new_df18['director_name'] +' ' + new_df18['genres']
new_df18.head()
```
