Here, we observe that we have movies from 1916 till 2016. The dataset contains most of the movies in between 1997 and 2016. We also have NaN values which we need to treat before proceeding ahead.

Also, we will first select the features that are important to us for recommmending movies on the OTT platform. All the information of each movie cannot be displayed in the suggestions. So, selecting the relevant features to display the useful information in recommendation for better decision making to user.

```python
# recommendation will be based on these features only
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
data.head()
```

```python
data.isna().sum()
```

We will replace the null values with the unknwon string as we don't have the relavant information for those cells.

```python
data['director_name']=data['director_name'].replace(np.nan,'unknown')
data['actor_1_name']=data['actor_1_name'].replace(np.nan,'unknown')
data['actor_2_name']=data['actor_2_name'].replace(np.nan,'unknown')
data['actor_3_name']=data['actor_3_name'].replace(np.nan,'unknown')
```

```python
data.isna().sum()
```

We notice a special character '|' as a separator in the genre column. We will replace this character with a blank space for simplicity.

```python
data['genres']=data['genres'].str.replace('|', ' ')
```

```python
data.head()
```

Now, we will lowercase the movie title for simplicity in further analysis.

```python
data['movie_title']=data['movie_title'].str.lower()
```

```python
data.head()
```

```python
data['movie_title'][21]
```

Here, we see some terminating character at the end of string of each movie title. We need to process this so as to avoid erros in further analysis. 

```python
data['movie_title']=data['movie_title'].apply(lambda x : x[:-1])
```

```python
#rechecking the movie column
data['movie_title'][35]
```

Now, the data looks clean for further analysis. We will save this first part of the data in a csv file and proceed with further movies data collection of years 2018, 2019 and 2020.
