```python
# we will convert the "genre", "cast" and "crew" column into a list. if we observe it carefully, we will realise that it's a list containing dictionary. 
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))  #the "literal_eval" helps to convert the string into a list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

ast.literal_eval function helps us to read the string containing python values. When we pass values to this function, it returns the actual type of the objects contained in it. Here, we see that these features contain list of dictionaries. So, it will return list as the data type and convert these columns into list.

```python
data.dtypes
```

Now, we observe that genres column contains list of dictionaries. Each dictionary has id of genre and name of genre. We are interested in fetching the names of each genre and append it into a new column with blank space as separator between them.

```python
#we will fetch the individual genres like adventure, action, sci-fi, etc. using the following function

def make_genresList(x):
    gen = []
    st = " " #will be used to separate the different values of genres with a blank space.
    for i in x:
        if i.get('name') == 'Science Fiction': #renaming Science Fiction to Sc-Fi for shorter name. All the other names are kept as it is.
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen)) #then we will join them together and return the values

#The get() method returns the value of the item with the specified key. We have dictionaries with key-pair values. We get the names of genres using the name as key and value as the genre name.
```

```python
#applying the function to genres column
data['genres_list']=data['genres'].map(lambda x: make_genresList(x))
```

```python
data.head()
```

Similarly, we will be fetching the names of the actors from the cast column and storing it in the new features.

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name']=data['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name']=data['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name']=data['cast'].map(lambda x: get_actor3(x))
```

```python
data.head()
```

```python
data['crew'][1]
```

Similarly, we will fetch the director name from the crew column. This column also contains list of dictionaries where each dictionary contains information about each crew member.

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name']=data['crew'].map(lambda x: get_directors(x))
```

```python
data.head()
```
